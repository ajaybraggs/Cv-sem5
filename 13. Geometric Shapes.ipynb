```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img_line=np.zeros((500,500,3),dtype="uint8")
cv2.line(img_line,(10,100),(100,100),(0,0,255),5)
plt.imshow(img_line)
plt.show()
```

```python
img_rec=np.zeros((500,500,3),dtype="uint8")
cv2.rectangle(img_rec,(30,30),(300,200),(0,255,0),5)
plt.imshow(img_rec)
plt.show()
```

```python
img_cir=np.zeros((500,500,3),dtype="uint8")
cv2.circle(img_cir,(200,200),80,(255,0,0),3)
plt.imshow(img_cir)
plt.show()
```

```python
img_text=np.zeros((500,500,3),dtype="uint8")
font=cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
cv2.putText(img_text,"AJAY BRAGGS",(100,200),font,1.0,(0,255,0),2,cv2.LINE_AA)
plt.imshow(img_text)
plt.show()
```
